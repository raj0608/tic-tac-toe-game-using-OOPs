```python
#importing necessary modules
import random 
### Global Variables

#variable to represent our board
board = ["_" for i in range(9)]

#game will be go on until false
game_on = True

current = "x"

#used to find the winner
winner = None

#tic tac toe  game function
def tttgame():
    display_board()
    #input for type of game i.e. Single player or multi player
    var1 = int(input("Choose '1' for 2 player game or '2' for vs computer: "))
    
    #in case of multiplayer
    if var1 == 1:
        
        name1 = input("Enter the Name of player x : ")
        name2 = input("Enter the Name of player o: ")
        
        while game_on:
            input_pos(current)
            result()
            switch_player()
        
        if winner == "x":
            print(name1+" Won :)")
            quit_game()
        elif winner == "o":
            print(name2+" Won :)")
            quit_game()
        elif winner == None:
            print("The game is Tie")
            quit_game()
    if var1 == 2:
        name1 = input("Enter name of player x: ")
        while game_on:
            if current == "x":
                input_pos(current)
            else:
                #function  call for computer's chance
                auto_pos()
            result()
            switch_player()
        
        if winner == "x":
            print(name1+" Won :)")
            quit_game()
        elif winner == "o":
            print("Computer won Won :)")
            quit_game()
        elif winner == None:
            print("The game is Tie")
            quit_game()
        

def display_board():
    print("\n")
    print(board[0] + " | " + board[1] + " | " + board[2] + "     1 | 2 | 3")
    print(board[3] + " | " + board[4] + " | " + board[5] + "     4 | 5 | 6")
    print(board[6] + " | " + board[7] + " | " + board[8] + "     7 | 8 | 9")
    print("\n")

#function to return coputer's position 
def auto_pos():
    pos = random.randint(1,9)
    if board[pos-1] == "_":
        board[pos-1] = "o"
    else:
        auto_pos()
    display_board()
    quit_now("o")
    
#function to handle a single turn
def input_pos(current):
    # Get position from player
    print(current + "'s turn.")
    pos = int(input("Choose a number between 1-9: "))
    if pos >= 1 and pos <= 9: 
        pos = pos -1
        if board[pos] == "_":
            #updating value in board
            board[pos] = current
    else:
        print("Please enter correct input")
        input_pos(current)
    display_board()
    quit_now(current)

    
#function to handle result        
def result():
    find_winner()
    find_if_tie()

#function to find winner
def find_winner():
    global winner
    # Check if there was a winner anywhere
    row_winner = check_rows()
    column_winner = check_columns()
    diagonal_winner = check_diagonals()
    # Get the winner
    if row_winner:
        winner = row_winner
    elif column_winner:
        winner = column_winner
    elif diagonal_winner:
        winner = diagonal_winner
    else:
        winner = None

#to check if rows satisfy condition to win
def check_rows():
    global game_on
    # Check if any of the rows have all the same value and not empty
    row_1 = board[0] == board[1] == board[2] != "_"
    row_2 = board[3] == board[4] == board[5] != "_"
    row_3 = board[6] == board[7] == board[8] != "_"
    # If any row does have a match, flag that there is a win
    if row_1 or row_2 or row_3:
        game_on = False
    # Return the winner
    if row_1:
        return board[0] 
    elif row_2:
        return board[3] 
    elif row_3:
        return board[6] 
    # Or return None if there was no winner
    else:
        return None

#to check if columns satisfy condition to win
def check_columns():
    global game_on
    # Check if any of the columns have all the same value (and is not empty)
    column_1 = board[0] == board[3] == board[6] != "_"
    column_2 = board[1] == board[4] == board[7] != "_"
    column_3 = board[2] == board[5] == board[8] != "_"
    # If any row does have a match, flag that there is a win
    if column_1 or column_2 or column_3:
        game_on = False
    # Return the winner
    if column_1:
        return board[0] 
    elif column_2:
        return board[1] 
    elif column_3:
        return board[2] 
    # Or return None if there was no winner
    else:
        return None    
    

#to check if diagonals satisfy condition to win
def check_diagonals():
    global game_on
    #Check if any of the columns have all the same value and is not empty
    diagonal_1 = board[0] == board[4] == board[8] != "_"
    diagonal_2 = board[2] == board[4] == board[6] != "_"
    # If any row does have a match, flag that there is a win
    if diagonal_1 or diagonal_2:
        game_on = False
    # Return the winner
    if diagonal_1:
        return board[0] 
    elif diagonal_2:
        return board[2]
    # Or return None if there was no winner
    else:
        return None
#function to find if result is tie    
def find_if_tie():
    global game_on
    # If board is full
    if "_" not in board:
        game_on = False
        return True
    # Else there is no tie
    else:
        return False
    
#function to change the player 
def switch_player():
    global current
    
    #interchanging value of current
    if current == "x":
        current = "o"
    else:
        current = "x"

#function to quit game during the game
def quit_now(current):
    global game_on
    global winner
    x , o = "x" , "o"
    var3 = input("Do you want to quit now(y/n): ")
    if var3 == "y":
        if current == "x": winner = o
            
        else: winner = x
        game_on = False
        
    else:
        return None 


#function to continue or quit at the end of the game
def quit_game():
    global board
    global current
    global game_on 
    var2 = input("to continue enter 'yes' and to quit enter 'no': ")
    if var2 == "yes":
        #here we will change to initial conditions true for main game function
        board = ["_" for i in range(9)]
        current = "x"
        game_on = True
        tttgame()
    else:
        print("GG")

#calling the function
tttgame()
```
